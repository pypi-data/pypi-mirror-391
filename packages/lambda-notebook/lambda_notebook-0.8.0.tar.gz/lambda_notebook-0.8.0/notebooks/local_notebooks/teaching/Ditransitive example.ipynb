## Ditransitive example
### Semantics 1 at JHU, 10/2/23

```python
nltk.Tree.fromstring(
    "(S (NP (N Alfonso)) (VP (V' (V introduced) (NP (N Joanna))) (PP (P to) (NP (N Henry)))))")
```

```python
%%lamb
||Alfonso|| = a
||Joanna|| = j
||Henry|| = h
```

```python
%%lamb
||introduce|| = L y_e : L z_e : L x_e: Introduce(x,y,z)
||to|| = L x_e : x
```

```python
nltk.Tree.fromstring("(V' introduced Joanna)")
```

```python
# "bottom-up" composition: use the * operator to compose smaller pieces into bigger pieces
(introduce * Joanna).tree()
```

```python
nltk.Tree.fromstring("(PP to Henry)")
```

```python
(to * Henry).tree()
```

```python
nltk.Tree.fromstring("(VP (V' introduce Joanna) (PP to Henry))")
```

```python
((introduce * Joanna) * (to * Henry)).tree()
```

```python
(Alfonso * ((introduce * Joanna) * (to * Henry))).tree()
```

```python
# note: argument is always shown after function in this approach to composition
lamb.display.default(style=lamb.display.DerivStyle.TREE)
(Alfonso * ((introduce * Joanna) * (to * Henry))).tree()
```

## Repeated PM example

```python
%%lamb
||small|| = L x_e : Small(x)
||gray|| = L x_e : Gray(x)
||spotted|| = L x_e : Spotted(x)
||cat|| = L x_e : Cat(x)
```

```python
(small * (gray * (spotted * cat)))
```

```python
# lamb.display.reset_default()
lamb.display.default(style=lamb.display.DerivStyle.PROOF)
(small * (gray * (spotted * cat))).tree()
```

```python
lamb.display.default(style=lamb.display.DerivStyle.TREE) # (note: small rendering issue with this)
(small * (gray * (spotted * cat))).tree()
```
