```python
from lamb.meta.core import TypedExpr, SyncatOpExpr

class DataOpExpr(SyncatOpExpr):
    arity = 1
    operator_style = False
    def __init__(self, data, name, typ=None, **kwargs):
        self.canonical_name = name
        self.data = data
        self.value = value
        typ, _ = self.type_constraint(value.type, typ, tcheck_args=False)
        super().__init__(self.data, typ=typ)

class DataExprFacade(lamb.meta.core.TypedExprFacade):
    def __getattr__(self, name):
        if name in self._e.fields:
            return self._e.get_value(name)
        else:
            return super().__getattr__(name)

def data_type(type_name, params):
    class DataValue(utils.Namespace):
        def __init__(self, **kwargs):
            self.fields = {x.op: x.type for x in params}
            # self.values = {x.op: params[x] for x in params}
            super().__init__(kwargs, {x.op: params[x] for x in params})

        def __setitem__(self, key, value):
            if key not in self.fields:
                raise ValueError(f"Invalid key {key} for data value")
            self.items[key] = value
    
        def __delitem__(self, key):
            raise NotImplementedError()
        
```
